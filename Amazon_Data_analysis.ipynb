**Steps to delete blank Columns**

```python
df.drop(['New','PendingS'],axis=1, inplace=True)
```

```python
df.info()
```

**Checking for Null Values**

```python
pd.isnull(df)
```

**Total Values of null**

```python
pd.isna(df).sum()
```

**Deleting Null Values**

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isna(df).sum()
```

**Changing Data Type**

```python
df['ship-postal-code']=df['ship-postal-code'].astype(int)
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Date'].dtype
```

```python
df.info()
```

**Renaming of Column**

```python
df.rename(columns={'Qty':'Quantity'},inplace=True)
```

```python
df.head()
```

```python
df.describe(include='object')
```

```python
df[['Quantity','Amount']].describe()
```

**EDA**

```python
ax=sns.countplot(x='Size',data=df)
```

**Adding Data Lebels to the bar**

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

**From the above graph we can see that the most popular Size is M-size**

```python
df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
sqty=df.groupby(['Size'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Size',y='Quantity',data=sqty)
```

**From the above graph we can see that most quantity sold is M-size.**

**Courier Status**

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Courier Status',data=df,hue='Status')
plt.show()
```

**Majority of the orders shipped through Courier**

```python
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
c_data=df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_data,bins=15,edgecolor='black',color='green')
plt.xticks(rotation=90)
plt.show()
```

**From the above graph we can see that T-shirt is the most sold Category**

```python
B2B_C=df['B2B'].value_counts()
plt.pie(B2B_C,labels=B2B_C.index,autopct='%1.1f%%')
plt.show()
```

**From the above chart we can see that maximum i.e 99.2% of buyers are retailers and 0.8% are B2B buyers**

```python
x_data=df['Category']
y_data=df['Size']
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.figure(figsize=(10,6))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='ship-state',data=df)
plt.xticks(rotation=90)
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.show()
```

**For Top 10 States**

```python
top_10_states=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.countplot(x='ship-state',data=df[df['ship-state'].isin(top_10_states.index)])
plt.xticks(rotation=45)
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.title('Distribution of States')
plt.show()
```

**From the above graph we can say the most of the buyers are from Maharashtra State.**
